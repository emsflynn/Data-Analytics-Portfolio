```python
import matplotlib.pyplot as plt
```

```python
X = [0,1,2,3,4,5,7,8,9,10]
Y = [20,25,35,50,10,12,20,40,70,110]
plt.plot(X,Y, marker='s', linestyle='--', color='magenta', lw=5)

#marker and linestyle and color has to be in '' 
#linewidth or lw
```

```python
#ALTERNATIVE using the fmt way
X = [0,1,2,3,4,5,7,8,9,10]
Y = [20,25,35,50,10,12,20,40,70,110]
fmt='s-.r'
plt.plot(X,Y,fmt,lw=5)
```

```python
#to print many, same way as print, shows all, plt.show before each one u want to print 

X = [0,1,2,3,4,5,7,8,9,10]
Y = [20,25,35,50,10,12,20,40,70,110]
plt.plot(X,Y, marker='s', linestyle='--', color='magenta', lw=5)

plt.show()

fmt='s-.r'
plt.plot(X,Y,fmt,lw=5)
```

```python
#date,man,woman

date = [2000,2002,2004,2006,2010,2012,2014,2016]
man = [200,191,187,198,173,181,192,185]
woman = [165,172,180,155,185,163,172,169]

heightsDf = {'Date' : date, 'Man' : man, 'Woman' : woman}
heightsDf
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(heightsDf)
df
```

```python
plt.plot(df['Date'], df['Man'], marker='s', linestyle='--', color='magenta', lw=5)
```

```python
fmt='s-.r'
plt.plot(df['Date'], df['Man'],fmt,lw=5)
```

```python
#THIS IS THE WAY TO CALL FROM A DICTIONARY WITH 2 COLUMNS, HAVE TO SEPERATE IT OUT
plt.plot(heightsDf["Date"], heightsDf["Man"], color = 'r')
plt.plot(heightsDf["Date"], heightsDf["Woman"], color = 'y')
plt.legend(["Man", "Woman"])

plt.xlabel('Date')
plt.ylabel('Height(cm)')
plt.title('Heights over time')
plt.legend(['Woman','Men'])
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot('Date', 'Woman', data=heightsDf)
ax2.plot('Date', 'Man', data=heightsDf)
ax1.set_xlabel('Date')
ax1.set_ylabel('Heigh(cm)')
ax1.set_title('Heights over time')
ax1.legend(['Woman'])

ax2.set_xlabel('Date')
ax2.set_ylabel('Heigh(cm)')
ax2.set_title('Heights over time')
ax2.legend(['Man'])
```
