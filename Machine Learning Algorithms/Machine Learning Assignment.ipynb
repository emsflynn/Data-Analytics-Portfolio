```python
import numpy as np
import pandas as pd
```

```python
inPath = '/Users/emsfl/Downloads/'
inpDf = pd.read_csv(inPath + 'HR_DB.csv', sep=',', header=0, index_col=None)
```

```python
inpDf
```

```python
inpDf.shape
```

```python
inpDf = inpDf.dropna(axis=0, how='any')
```

```python
inpDf
```

```python
inpDf = pd.get_dummies(inpDf, columns=['division'], prefix='Dpt')
inpDf
```

```python
#LABEL ENCODING

inpDf['wage'].unique()
wageDict = {'low': 1, 'medium': 2, 'high': 3}
inpDf['wage'] = inpDf['wage'].map(wageDict)
```

```python
print(inpDf['recentPromotion'].skew())
print(inpDf['recentPromotion'].kurt())

#If normally distributed : skewness would be between -2 and 2 and kustrosis would be between -7 and 7 

#THERE, IT IS NOT NORMALLY DISTRIBUTED
```

```python
#standardizing variable
inpDf['avgMtlyHrs'] = (inpDf['avgMtlyHrs'] - inpDf['avgMtlyHrs'].mean()) / inpDf['avgMtlyHrs'].std()
```

```python
inpDf['avgMtlyHrs']
```

```python
yDf = inpDf['resigned']
xDf = inpDf.drop(columns='resigned')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(xDf, yDf, test_size=0.3, random_state=0)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#set algorithm parameters
clf = svm.SVC(C=100, kernel='rbf', gamma=1, random_state=0)
```

```python
#Fit the data
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
