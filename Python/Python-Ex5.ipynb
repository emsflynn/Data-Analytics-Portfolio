```python
outliers = subDF[abs(z_scores) > z_score_threshold]
```

```python
upper_bound = subDF['can_off_dis'].mean() + z_score_threshold * subDF['can_off_dis'].std()
lower_bound = subDF['can_off_dis'].mean() - z_score_threshold * subDF['can_off_dis'].std()
```

```python
#HIS WAY - -Identify the upper and lower bounds using the Inter Quartile Range (IQR) method for the ‘can_off_dis’
#variable


lowZVal = subDF['can_off_dis'].mean() - 3 * subDF['can_off_dis'].std()
upZVal = subDF['can_off_dis'].mean() + 3 * subDF['can_off_dis'].std()
```

```python
#HIS WAY - only keep the rows in ‘subDF’ having values for ‘can_off_dis’
#within the boundaries

noOutlrZDf = subDF[(subDF['can_off_dis']>lowZVal) & (subDF['can_off_dis']<upZVal)]
```

```python
#In a DataFrame called ‘noOutlrZDf’, only keep the rows in ‘subDF’ having values for ‘can_off_dis’
#within the boundaries

within_bounds = (abs(z_scores) <= z_score_threshold)
noOutlrZDf = subDF[within_bounds]
```

```python
noOutlrZDf
```

```python
#Identify the upper and lower bounds using the Inter Quartile Range (IQR) method for the ‘can_off_dis’
#variable


Q1 = subDF['can_off_dis'].quantile(0.25)
Q3 = subDF['can_off_dis'].quantile(0.75)

IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

```

```python
#In a DataFrame called ‘noOutlrIqrDf’, only keep the rows in ‘subDF’ having values for ‘can_off_dis’
#within the boundaries


within_bounds = (subDF['can_off_dis'] >= lower_bound) & (subDF['can_off_dis'] <= upper_bound)

noOutlrIQRDf = subDF[within_bounds]
```

```python
#a function to convert string to numeric while replacing the '$' to '' and the ',' to '' (use replace)

def convert_to_numeric(value):
    if isinstance(value,str):
        try:
            return pd.to_numeric(value.replace('$', '').replace(',',''))
        except ValueError:
            return value
    else:
        return value
```

```python
#Apply function to each column

for col in inpDf.columns:
    inpDf[col] = inpDf[col].apply(convert_to_numeric)
    
print(inpDf['ind_ite_con'])
```

```python
import pandas as pd
```

```python
#Convert dates in string format of 'cov_sta_dat' and 'cov_end_dat' columns to a datetime format using
#to_datetime (if not we cannot compute the difference for these dates)

inpDf['cov_sta_dat'] = pd.to_datetime(inpDf['cov_sta_dat'])
inpDf['cov_end_dat'] = pd.to_datetime(inpDf['cov_end_dat'])
```

```python
print(inpDf['cov_sta_dat'])
print(inpDf['cov_end_dat'])
```

```python
#-In a new column called ‘campaign_duration’, compute the differences between the end dates and the
#start dates and remove the days unit

inpDf['campaign_duration'] = (inpDf['cov_end_dat'] - inpDf['cov_sta_dat']).dt.days
```

```python
inpDf['campaign_duration']
```

```python
#Remove the 'cov_sta_dat' and 'cov_end_dat'

inpDf.drop(columns=['cov_sta_dat', 'cov_end_dat'], inplace=True)
```

```python
inpDf
```

```python
#In ‘inpDf’, set as missing value the negative values in 'campaign_duration'

inpDf.loc[inpDf['campaign_duration']<0, 'campaign_duration'] = np.nan
```

```python
inpDf
```

```python
#Use one-hot encoding to convert ‘can_par_aff’ (which is a nominal value, there is no order) to dummy
#variables using the prefix ‘party’

inpDf = pd.get_dummies(inpDf, columns=['can_par_aff'], prefix='party')
```

```python
#We consider that 'can_inc_cha_ope_sea' is an ordinal variable (at least for this example). As it is
#ordinal, we use label encoding

#Display all the possible values for this variable (use unique)

inpDf['can_inc_cha_ope_sea'].unique()
```

```python
#Create the encoding dictionary

encoding_dict = {'OPEN' : 0, 'CHALLENGER' : 1, 'INCUMBENT' : 2, np.nan:np.nan}
```

```python
print(encoding_dict)
```

```python
#Then apply it to 'can_inc_cha_ope_sea' using pandas and store it in a
#‘'can_inc_cha_ope_sea_encoded_pd’ variable.

inpDf['can_inc_cha_ope_sea_encoded_pd'] = inpDf['can_inc_cha_ope_sea'].map(encoding_dict)
```
