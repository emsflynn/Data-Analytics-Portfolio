```python
import numpy as np
```

```python
#Create a dictionary having the 3 following keys: Years, Women and Men. For the items, the dates
#should be ordered between 2000 and 4064 (with a step of 2). Then using NumPy randint, create
#heights values for women and men respectively between 145 & 200 and between 155 & 215.
```

```python
women_heights = np.random.randint(145, 201, size=len(range(2000, 4065, 2)))
men_heights = np.random.randint(155, 216, size=len(range(2000, 4065, 2)))
```

```python
my_dict = {'Years': list(range(2000, 4065, 2)),
           'Women': list(women_heights),
           'Men': list(men_heights)}
```

```python
print(my_dict)
```

```python
import pandas as pd
```

```python
#Convert the dictionary to a dataframe
```

```python
df = pd.DataFrame(my_dict)
df
```

```python
#Plot the two lines for women and men on the same graph and include titles for the graph and for both
#axis, and a legend. Additionnally, the women line must be in green and the men line in yellow
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Years"], df["Men"], color = 'g')
plt.plot(df["Years"], df["Women"], color = 'y')
plt.legend(["Men", "Women"])

plt.xlabel('Years')
plt.ylabel('Height(cm)')
plt.title('Heights over time')
plt.legend(['Woman','Men'])

```

```python
# Count the missing data
print(inpDf.isna().sum())
# Drop the nan
inpDf = inpDf.dropna()
# Check that there are no missing data
inpDf.isna()
```

```python
# Compute the standardized values
inpDf.loc[:, 'StdzVal'] = (inpDf['Val'] - inpDf['Val'].mean()) / inpDf['Val'].std()
print(inpDf['StdzVal'])
```

```python
#Using describe, skew and kurtosis, conclude whether the data can be considered as Normallydistributed
print('Skewness: ' + str(inpDf['StdzVal'].skew()))
print('Excess Kurtosis: ' + str(inpDf['StdzVal'].kurtosis()))

#The values are respectivwely within -2 and 2, and between -7 and 7 thus we consider this variable as Normal.
```
